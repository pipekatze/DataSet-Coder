```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que ya tienes un DataFrame llamado 'df'
# Si no tienes 'df', asegúrate de cargar tus datos correctamente en un DataFrame.

# Filtra solo las columnas numéricas
columnas_numericas = df.select_dtypes(include=['number']).columns

# Crea un nuevo DataFrame con las columnas numéricas
df_numericas = df[columnas_numericas]

# Calcula la matriz de correlación usando el método de Pearson
matriz_correlacion_pearson = df_numericas.corr()

# Transforma la matriz de correlación en un formato adecuado para el gráfico de barras horizontales
correlaciones = matriz_correlacion_pearson.unstack().sort_values(ascending=False)

# Filtra las correlaciones duplicadas y las correlaciones de las variables consigo mismas
correlaciones = correlaciones[correlaciones != 1]

# Selecciona las 20 correlaciones más significativas
correlaciones_top20 = correlaciones.head(20)

# Graficar las 20 correlaciones más significativas utilizando un gráfico de barras horizontales
plt.figure(figsize=(10, 8))
correlaciones_top20.plot(kind='barh', color='skyblue')
plt.xlabel('Correlación de Pearson')
plt.ylabel('Pares de variables')
plt.title('20 correlaciones de Pearson más significativas entre pares de variables numéricas')
plt.show()
```
